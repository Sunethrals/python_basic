```python
import datetime
class bank:
    bname="SBI"
    branch="Tenali"
    pincode=522201
    ifsc="SBI535SU"
    toll_free_no=9440831532
    roi=6
    no_cust=0
    cust_details={}
    cust_transactions={}
    def __init__(self,name,age,email,phone,address,adhaar,pancard,dob,gender,pin,bal=0):
        self.name=name
        self.age=age
        self.email=email
        self.phone=phone
        self.address=address
        self.adhaar=adhaar
        self.pancard=pancard
        self.dob=dob
        self.gender=gender
        self.pin=pin
        self.bal=bal
        self.no_cust_inc()
        self.account_no=100+self.no_cust
        self.concat_cust_details(self.account_no,self)
    @classmethod
    def no_cust_inc(cls):
        cls.no_cust+=1
    @classmethod
    def concat_cust_details(cls,acc_no,obj):
        cls.cust_details[acc_no]=obj
    @classmethod
    def change_roi(cls):
        cls.roi=int(input("enter the roi:"))
        print("roi is changed")
    @classmethod
    def check_bal(cls):
        acc_no=int(input("enter the account no:"))
        pin=int(input("enter the account pin no"))
        for i in cls.cust_details:
            res=cls.cust_details[i]
            if res.account_no==acc_no and res.pin==pin:
                print(res.account_no,"balance is:",res.bal)
                return
        else:
            print("invalid account_no or pin")
    @classmethod
    def deposit(cls,count=0):
        if count == 3:
            return
        acc_no=int(input("enter the account no:"))
        pin=int(input("enter the account pin no"))
        for i in cls.cust_details:
            res=cls.cust_details[i]
            
            amt=int(input("enter the deposit amount:"))
            if res.account_no==acc_no and res.pin==pin :
                if amt>=50:
                    res.bal+=amt
                # self.success()
                # self.concat_cust_tran(self.account_no,{'date':datetime.datetime.now(), 'type':'deposit','amount':amt,'bal':self.bal})
                    return
                else:
                    print("invalid account_no or pin")
            else:
                print("amount should be minimum 50rs")
                return res.deposit(count+1)
    @classmethod
    def withdraw(cls,count=0):
        acc_no=int(input("enter the account no:"))
        pin=int(input("enter the account pin no"))
        if count == 3:
            return
        for i in cls.cust_details:
            res=cls.cust_details[i]
            amt=int(input("enter the withdraw amount:"))
            if res.account_no==acc_no and res.pin==pin :
                if amt>=100 and res.bal>=amt:
                    res.bal-=amt
                    return
                else:
                    print("invalid account_no or pin")
            else:
                print("amount should be minimum 100 rs")
                return res.withdraw(count+1)

    @classmethod
    def pin_change(cls):
        acc_no=int(input("enter the account no:"))
        # pin=int(input("enter the account pin no"))
        old=int(input("enter the new pin:"))
        for i in cls.cust_details:
            res=cls.cust_details[i]
            if res.account_no==acc_no and res.pin==old:
                res.pin=int(input("enter the new pin:"))
                print("pin changed")
                return
        else:
            print("invalid account_no or pin")

    def mini_statement(self):
        print('Date'.center(20),'Type'.center(10),'Amount'.center(10),'Balance'.center(10),sep='|')
        print(80*'-')
            
        for i in self.cust_transactions[self.account_no]:
            print(str(i['date']).center(20),str(i['type']).center(10),str(i['amount']).center(10),str(i['bal']).center(10),sep='|')
            print(80*'-')

    @classmethod
    def cust_update(cls):
        acc_no=int(input("enter the account no:"))
        pin=int(input("enter the account pin no"))
        for i in cls.cust_details:
            res=cls.cust_details[i]
            if res.account_no==acc_no and res.pin==pin:
                n=int(input("choose the option:\n1.Name\n2.Phone\n3.email\n4.address"))
                match n:
                    case 1:
                        res.name=input("enter the name:")
                    case 2:
                        res.phone=int(input("enter the phone:"))
                    case 3:
                        res.email=input("enter the email")
                    case 4:
                        res.address=input("enter the address")
                    case _:
                        return
                return 
            else:
                print("invalid account_no or pin")

    @classmethod
    def account_details(cls):
        acc_no=int(input("enter the account no:"))
        pin=int(input("enter the account pin no"))
        for i in cls.cust_details:
            res=cls.cust_details[i]
            if res.account_no==acc_no and res.pin==pin:
                print("Account No:",res.account_no)
                print("Name:",res.name)
                print("Age:",res.age)
                print("Email:",res.email)
                print("Phone:",res.phone)
                print("Balance:",res.bal)
                print("IFSC code:",res.ifsc)
                print("Pan:",res.pancard)
                print("Adhaar:",res.adhaar)
                print("Address:",res.address)
            else:
                print("invalid account_no or pin")
    @classmethod
    def concat_cust_tran(cls,acc_no,data):
        if acc_no in cls.cust_transactions:
            cls.cust_transactions[acc_no]+=[data]
            print(acc_no,data)
        else:
            cls.cust_transactions[acc_no]=[data]
            print(acc_no,data)

    def transfer(self):
        print("Your current Balance:",self.bal)
        acc_no=int(input("Enter The Benificiary Accno:"))
        ifsc_code=input("Enter The Benificiary IFSC:")
        if acc_no in self.cust_details and ifsc_code==self.cust_details[acc_no].ifsc:
            amt=int(input("Enter The Transfer Amount:"))
            if self.bal>=amt:
                other=self.cust_details[acc_no]
                other.bal+=amt
                self.bal-=amt
                self.concat_cust_tran(self.account_no,{'date':datetime.datetime.now(), 'type':'Transfer(w)','amount':amt,'bal':self.bal})
                self.concat_cust_tran(other.account_no,{'date':datetime.datetime.now(), 'type':'Transfer(D)','amount':amt,'bal':other.bal})

                self.success()
                print("After Transfer Current Balance:",self.bal)
                return
            else:
                self.failed()
                print("Insufficient Balance")
        else:
            print("Invalid Account Number or IFSC Code")
        return

    @classmethod
    def home(cls,self):
        n=int(input("CHOOSE THE OPTIONS:\n0.ACCOUNT DETAILS\n1.CHECK BAL\n2.DEPOSIT\n3.WITHDRAW\n4.MINI STATEMENT\n5.UPDATE\n6.PIN CHANGE\n7.CHECK ROI\n8.TRANSFER\n9.EXIT"))
        print("*"*200)
        match n:
            case 0:
                self.account_details()
            case 1:
                self.check_bal()
            case 2:
                self.deposit()
            case 3:
                self.withdraw()
            case 4:
                self.mini_statement()
            case 5:
                self.cust_update()
            case 6:
                self.pin_change()
            case 7:
                print("Today RATE OF INTEREST",self.roi)

            case 8:
                self.transfer()
            case _:
                return
        return
obj=bank("sunethra",22,"sunethrals9@gmail.com",7396929629,"tenali",250656201235,"SUN123M","21-12-2001","female",12345,2000)
print(obj.__dict__)
print(bank.__dict__)
def main(cls):
    acc_no=int(input("Enter the Account_no:"))
    pin=int(input("Enter the Pin_no:"))
    if acc_no in cls.cust_details and cls.cust_details[acc_no].pin==pin:
        self=cls.cust_details[acc_no]
        cls.home(self)
    else:
        print("Invalid account_no or pin")
# main(bank)
# bank.change_roi()
# obj.change_roi()
# bank.check_bal()
# print(bank.cust_details)
# print(bank.__dict__)
# obj.deposit()
# bank.withdraw()
# obj.cust_update()
bank.cust_update()
# obj.transfer()
# obj.mini_statement()
# bank.pin_change()

print(obj.__dict__)
```
