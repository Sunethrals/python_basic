```python
class employee:
    company_name='spiders'
    company_id='123'
    company_address='Hyderabad'
    emp_no=0
    emp_details={}
    def __init__(self,ename,phone,salary,address,domain,emailid):
        self.ename=ename
        self.phone=phone
        self.salary=salary
        self.address=address
        self.domain=domain
        self.emailid=emailid
        self.inc_empno()
        self.emp_id=0+self.emp_no
        self.concat_emp_details(self.emp_id,self)
    
    @classmethod
    def inc_empno(cls):
        cls.emp_no+=1

    @classmethod
    def concat_emp_details(cls,id,self):
        cls.emp_details[id]=self.__dict__

    @classmethod
    def display_emp_details(cls):
        for i in cls.emp_details:
            print(cls.emp_details[i])
            cls.suscess()
    
    @classmethod
    def hike_sal(cls):
        for i in cls.emp_details:
            # cls.emp_details['salary']=50000
            cls.complete()
            print(cls.emp_details[i])
            cls.emp_details[i]['salary']=cls.emp_details[i]['salary']+cls.emp_details[i]['salary']*4/100
            # print(cls.emp_details[i]['salary'])
            cls.after()
            print(cls.emp_details[i])

    @classmethod
    def domain(cls):
        m=str(input("enter the domain"))
        # print(m)
        for i in cls.emp_details:
            if cls.emp_details[i]['domain'] == m:
                print(cls.emp_details[i])

    @classmethod
    def result(cls):
        n=eval(input("enter your choice : 1.display emp details\n 2.hike of salay \n 3. particular domain \n4.exit"))
        match n:
            case 1:
                cls.display_emp_details()
            case 2:
                cls.hike_sal()
            case 3:
                cls.domain()
            case _:
                return
    
    @staticmethod
    def suscess():
        print("DISPLAYED ALL THE DETAILS")
    @staticmethod
    def complete():
        print("before hike")
    @staticmethod
    def after():
        print("after hike")
obj1=employee('sunethra',7396929629,30000,'tenali','development','sunethrals9@gamil.com')
obj2=employee('lakshmi',7416171944,45000,'hyderabad','testing','laxsmi@gmail.com')
# obj1.__dict__
employee.__dict__
# employee.display_emp_details()
# employee.hike_sal()
# employee.emp_details
# employee.result()
# employee.domain()
```
