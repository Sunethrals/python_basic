```python
st='hello   hai hai'
res=""
s=0
while s<len(st):
    if st[s]==" ":
        res+=st[s]
        if st[s+1]!=" ":
            res+=str(chr(ord(st[s+1])-32))
        elif st[s+2]!=" ":
            res+=str(chr(ord(st[s+2])-32))
        elif st[s+3]!=" ":
            res+=str(chr(ord(st[s+3])-32))
    else:
        res+=st[s]
    s+=1
print(res)
```
